```python
class RobinNet:
    def __init__(self):
        self.l1 = nn.Linear(4,16)
        self.l2 = nn.Linear(16,16)
        self.l3 = nn.Linear(16,3)
    def __call__(self, x):
        x = self.l1(x)
        x = x.relu()
        x = self.l2(x)
        x = x.relu()
        x = self.l3(x)
        return x
```
